  **역공학, 재공학, 재사용**

#  **협약에 의한 설계**
  - **선행조건** : 컴포넌트 오퍼레이션 사용 전에 참이 되어야 할 조건
  - **결과조건** : 사용 후 만족되어야 할 조건(결과물)
  - **불변조건** : 오퍼레이션이 실행되는 동안 항상 만족되어야 할 조건


소프트웨어 아키텍쳐 4+1 뷰

- 4는 만드는 사람 뷰

- 1은 사용자 뷰

- 논리뷰, 구현뷰, 프로세스뷰, 배치뷰, 유스케이스뷰(사용자)


# **모델-뷰-컨트롤러 패턴(MVC Pattern)**
- 3개의 컴포넌트가 각각의 역할을 갖고 사용자에게 서비스를 제공
- 자료의 저장, 제어, 표현 기능을 분리
  - **모델**
    
    자료를 저장하고 보관 (저장)
  - **뷰**

    사용자에게 결과를 표시 (표현)
  - **컨트롤러**

    사용자로부터 입력을 받아 연산을 처리 (제어)



# **UML**

 - 사물
 - 관계
    - **일반화 관계**(상속) : 한 클래스가 다른 클래스를 포함하는 상위 개념일 때의 관계
    
    (ex] 가전제품이라는 부모클래스 안에 세탁기, TV, 식기세척기 같은 자식 클래스가 있음 )
    - **연관관계** : 2개 이상 사물이 서로 관련된 관계(오랜시간)
    - **의존관계** : 연관관계와 비슷 (짧은시간), 한 클래스가 다른 클래스를 오퍼레이션의 매개변수로 사용하는 경우
    - 실체화 관계 : 한 객체가 다른 객체에게 오퍼레이션을 수행하도록 지정
    
    (ex] 리모콘 틀에서 +/- 버튼이 에어컨은 온도 버튼, TV리모콘은 소리 버튼 )
    - (집합)- **집약관계** : 전체 객체의 라이프타임과 부분객체의 라이프타임은 독립적,
    
    전체 객체가 사라져도 부분 객체는 존재
      
      (ex] 불고기 - 간장, 설탕, 소고기, 당근, 시금치 )
    - (집합)- **합성관계** : 전체 객체의 라이프타임과 부분객체의 라이프타임은 의존적,
    
    전체 객체가 사라지면 부분 객체도 없어짐
    
    (ex] 책상 - 상판, 책상다리, 나사)
